```python
# create a to do list to keep track of tasks
# create a list to store tasks
tasks = []
# create a loop to keep the program running
while True:
    # print out a menu of options
    print("To Do List")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task Complete")
    print("4. Exit")
    # ask the user for their choice
    choice = input("Enter your choice: ")
    # if the user chooses 1, add a task
    if choice == "1":
        task = input("Enter a task: ")
        tasks.append(task)
    # if the user chooses 2, view tasks
    elif choice == "2":
        for i, task in enumerate(tasks):
            print(f"{i + 1}. {task}")
    # if the user chooses 3, mark a task complete
    elif choice == "3":
        for i, task in enumerate(tasks):
            print(f"{i + 1}. {task}")
        task_number = int(input("Enter the number of the task you completed: "))
        tasks.pop(task_number - 1)
    # if the user chooses 4, exit the program
    elif choice == "4":
        break
    # if the user chooses an invalid option, print an error message
    else:
        print("Invalid choice. Please try again.")
    # print a blank line
    print()
```

```python
# colleting users feedback , counting specfoc feedbacks, postive feed backs count and negative count
feedbacks = []
positive_feedbacks = 0
negative_feedbacks = 0
while True:
    # print out a menu of options
    print("Feedback")
    print("1. Add Feedback")
    print("2. View Feedback")
    print("3. View Positive Feedback")
    print("4. View Negative Feedback")
    print("5. Exit")
    # ask the user for their choice
    choice = input("Enter your choice: ")
    # if the user chooses 1, add a feedback
    if choice == "1":
        feedback = input("Enter a feedback: ")
        feedbacks.append(feedback)
    # if the user chooses 2, view feedback
    elif choice == "2":
        for i, feedback in enumerate(feedbacks):
            print(f"{i + 1}. {feedback}")
    # if the user chooses 3, view positive feedback
    elif choice == "3":
        for feedback in feedbacks:
            if "good" in feedback:
                print(feedback)
                positive_feedbacks += 1
    # if the user chooses 4, view negative feedback
    elif choice == "4":
        for feedback in feedbacks:
            if "bad" in feedback:
                print(feedback)
                negative_feedbacks += 1
    # if the user chooses 5, exit the program
    elif choice == "5":
        break
    # if the user chooses an invalid option, print an error message
    else:
        print("Invalid choice. Please try again.")
    # print a blank line
    print()
```
